```python
%reload_ext autoreload
%autoreload 2

import numpy as np
import matplotlib.pyplot as plt

from main import ClifWalking, Agent
```

# Standard moves

```python
env = ClifWalking()
agent = Agent(action_size=len(env.actions), alpha=0.1)

rewards_sarsa = agent.sarsa(env, 1000)
opt_trajectory_sarsa = agent.get_optimal_trajectory(env)

rewards_qlearning = agent.qlearning(env, 1000)
opt_trajectory_qlearning = agent.get_optimal_trajectory(env)
```

```python
fig, ax = plt.subplots(1, 1, figsize=(8, 4))

ax.plot(env.initial_state[1], env.initial_state[0], "og", markersize=20)
ax.plot(env.terminal_state[1], env.terminal_state[0], "or", markersize=20)
ax.plot([-111], [-111], "ob"), ax.plot([-111], [-111], "o", color="aqua")

for i in opt_trajectory_sarsa:
    ax.plot(i[1], i[0], "ob")

for i in opt_trajectory_qlearning:
    ax.plot(i[1], i[0], "o", color="aqua")

ax.legend(["Start", "Goal", "SARSA", "QLEARNING"])
ax.set_xlim([0 - 1, 10 + 1])
ax.set_ylim([0 - 1, 7 + 1])
ax.invert_yaxis()
fig.tight_layout()
```
