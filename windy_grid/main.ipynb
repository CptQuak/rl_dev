```python
%reload_ext autoreload
%autoreload 2

import numpy as np
import matplotlib.pyplot as plt

from main import WindyGridWorld, Agent
```

# Standard moves

```python
env = WindyGridWorld()
agent = Agent(action_size=len(env.actions))

agent.sarsa(env, 200)
```

```python
opt_trajectory = agent.get_optimal_trajectory(env)
fig, ax = plt.subplots(1, 1, figsize=(8, 4))

ax.plot(env.initial_state[1], env.initial_state[0], "og", markersize=20)
ax.plot(env.terminal_state[1], env.terminal_state[0], "or", markersize=20)

for i in opt_trajectory:
    ax.plot(i[1], i[0], "ob")

ax.set_xlim([0 - 1, 9 + 1])
ax.set_ylim([0 - 1, 7 + 1])
ax.invert_yaxis()
fig.tight_layout()
```
